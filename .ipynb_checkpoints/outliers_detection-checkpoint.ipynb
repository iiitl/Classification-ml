```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataclean_in-vehicle-coupon-recommendation.csv')
df.head()
```

```python
print(df.shape)
df.info()
```

```python
cols = ['temperature', 'has_children', 'toCoupon_GEQ5min', 'toCoupon_GEQ15min', 'toCoupon_GEQ25min' ,'direction_same' , 'direction_opp']

for col in cols:
    print(f"Unique values present in - {col}: {df[col].unique()}")
```

```python
cols = ['temperature', 'has_children', 'toCoupon_GEQ5min', 'toCoupon_GEQ15min']

fig, axes = plt.subplots(nrows=1, ncols=len(cols), figsize=(16, 4))
for i, col in enumerate(cols):
    sns.boxplot(y=df[col], ax=axes[i])
    axes[i].set_title(f'Original {col}')
plt.tight_layout()
plt.show()
```

```python
cols = ['toCoupon_GEQ25min' ,'direction_same' , 'direction_opp']

fig, axes = plt.subplots(nrows=1, ncols=len(cols), figsize=(16, 4))
for i, col in enumerate(cols):
    sns.boxplot(y=df[col], ax=axes[i])
    axes[i].set_title(f'Original {col}')
plt.tight_layout()
plt.show()
```

```python
cols = ['temperature', 'has_children', 'toCoupon_GEQ5min', 'toCoupon_GEQ15min']

fig, axes = plt.subplots(nrows=1, ncols=len(cols), figsize=(16, 4))
for i, col in enumerate(cols):
    sns.histplot(df[col], ax=axes[i])  # Use sns.histplot for histograms
    axes[i].set_title(f'{col}')
plt.tight_layout()
plt.show()
```

```python
cols = ['temperature', 'has_children', 'toCoupon_GEQ5min', 'toCoupon_GEQ15min']

fig, axes = plt.subplots(nrows=1, ncols=len(cols), figsize=(16, 4))
for i, col in enumerate(cols):
    sns.histplot(df[col], ax=axes[i])  # Use sns.histplot for histograms
    axes[i].set_title(f'{col}')
plt.tight_layout()
plt.show()
```

```python
X1 = df['temperature'].quantile(0.25)
X3 = df['temperature'].quantile(0.75)
IQR = X3-X1
lower_bound = X1 - 1.5 * IQR  
upper_bound = X3 + 1.5 * IQR

df['temperature'] = (df['temperature'] - lower_bound) / (upper_bound - lower_bound)
```
