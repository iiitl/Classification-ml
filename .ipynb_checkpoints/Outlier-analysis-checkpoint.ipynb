## Visualize outliers 

### Box Plots

```python
cols = ['temperature', 'has_children', 'toCoupon_GEQ5min', 'toCoupon_GEQ15min']

fig, axes = plt.subplots(nrows=1, ncols=len(cols), figsize=(16, 4))
for i, col in enumerate(cols):
    sns.boxplot(y=data[col], ax=axes[i])
    axes[i].set_title(f'Original {col}')
plt.tight_layout()
plt.show()
```

```python
cols = ['toCoupon_GEQ25min' ,'direction_same' , 'direction_opp']

fig, axes = plt.subplots(nrows=1, ncols=len(cols), figsize=(16, 4))
for i, col in enumerate(cols):
    sns.boxplot(y=data[col], ax=axes[i])
    axes[i].set_title(f'Original {col}')
plt.tight_layout()
plt.show()
```

## Histograms

```python
cols = ['temperature', 'has_children', 'toCoupon_GEQ5min', 'toCoupon_GEQ15min']

fig, axes = plt.subplots(nrows=1, ncols=len(cols), figsize=(16, 4))
for i, col in enumerate(cols):
    sns.histplot(data[col], ax=axes[i])  # Use sns.histplot for histograms
    axes[i].set_title(f'{col}')
plt.tight_layout()
plt.show()
```

# Conclusion

### > All the Numeric Columns except 'temperature' contains binary or single unique values and they dont need removal or transformation of outliers

### > For the Column 'temperature' the values are not that scattered and removing may cause loss of valuable data instead it capping or scaling will be better options also as here the extremes arent too high or too low i would prefer transforming or scaling the data in column 'temperature'. 


# Transforming 

```python
Q1 = data['temperature'].quantile(0.25)
Q3 = data['temperature'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR  
upper_bound = Q3 + 1.5 * IQR

data['temperature'] = (data['temperature'] - lower_bound) / (upper_bound - lower_bound)
```
