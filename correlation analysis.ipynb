```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("in-vehicle-coupon-recommendation.csv")
```

Selecting numerical features


```python
numerical_features = df.select_dtypes(include=['int64'])
```

Compute and create correlation matrix

```python

corr_matrix = numerical_features.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

# Observations
1. 'toCoupon_GEQ15min' and 'toCoupon_GEQ25min' are highly correlated, we can keep only one of them
2. 'direction_same' and 'direction_opp' are perfectly negatively correlated, one should be removed
