```python
import pandas as pd
```

```python
df=pd.read_csv('iris.csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
df.head()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
x=df.drop(columns=['Species'])
y=df['Species']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.svm import SVC
```

BEFORE GRIDSEARCHCV

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=200),
    "Random Forest": RandomForestClassifier(),
    "SVC" : SVC()
}
```

```python
def evaluate(model,x_train,y_train,x_test,y_test):
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    accuracy=accuracy_score(y_test,y_pred)
    precision=precision_score(y_test,y_pred,average='weighted')
    recall=recall_score(y_test,y_pred,average='weighted')
    f1=f1_score(y_test,y_pred,average='weighted')

    return accuracy,precision,recall,f1
```

```python
ans={}
for name,model in models.items():
    accuracy,precision,recall,f1=evaluate(model,x_train,y_train,x_test,y_test)
    ans[name]={'accuracy':accuracy,'precision':precision,'recall':recall,'f1':f1}
```

```python
for key,value in ans.items():
    print(key+ ":" )
    print(value)

```

AFTER GRIDSEARCHCV

```python
param_grids = {
    "Logistic Regression": {
        "C": [0.01, 0.1, 1, 10, 100],
        "max_iter": [200, 300, 400]
    },
    "Random Forest": {
        "n_estimators": [50, 100, 200],
        "max_depth": [None, 10, 20],
        "min_samples_split": [2, 5, 10]
    },
    "SVC": {
        "C": [0.1, 1, 10],
        "kernel": ['linear', 'rbf', 'poly']
    }
}
```

```python
def grid_search_cv(model, param_grid, X_train, y_train):
    search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    search.fit(X_train, y_train)

    best_model = search.best_estimator_
    best_params = search.best_params_

    return best_model, best_params
```

```python
ansgscv = {}
```

```python
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    return accuracy, precision, recall, f1
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
for name, model in models.items():

    best_model, best_params = grid_search_cv(model, param_grids[name], x_train, y_train)

    accuracy, precision, recall, f1 = evaluate(best_model, x_train, y_train, x_test, y_test)

    ansgscv[name] = {
        'best_params': best_params,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,

    }
```

```python
for name,value in ansgscv.items():
    print(name+ ":")
    print(value)
```
