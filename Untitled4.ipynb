```python
df_cl=df.drop("car",axis=1)
df_cl.shape
```

```python
df_cl.loc[(df_cl['Bar'].isnull())&(df_cl['CarryAway']=='never'),'Bar']='never'
```

```python
import numpy as np

bar_values = df_cl['Bar'].dropna().unique()
carryaway_values = df_cl['CarryAway'].dropna().unique()
def fill_bar(row):
    if pd.isna(row['Bar']) and row['CarryAway'] != 'never':
        return np.random.choice(bar_values)
    return row['Bar']

def fill_carryaway(row):
    if pd.isna(row['CarryAway']):
        return np.random.choice(carryaway_values)
    return row['CarryAway']
df_cl['Bar'] = df_cl.apply(fill_bar, axis=1)
df_cl['CarryAway'] = df_cl.apply(fill_carryaway, axis=1)
miss=df_cl.isnull().sum()
print(miss)
```

```python
import pandas as pd
import numpy as np

df=pd.read_csv("in-vehicle-coupon-recommendation.csv")
print(df[['RestaurantLessThan20','Restaurant20To50']])
```

```python
import pandas as pd

result = df_cl.groupby('RestaurantLessThan20')['Restaurant20To50'].unique()
result
```

```python
data = {'RestaurantLessThan20': ['1~3','4~8','gt8','less1','never','4~8','gt8',None],
        'Restaurant20To50':['never',None,'gt8',None,None,'4~8',None,'1~3']}
value_mapping = {
    '1~3': ['never','less1','1~3','4~8',np.nan],
    '4~8': ['1~3','less1','never','4~8',np.nan,'gt8'],
    'gt8': ['gt8','1~3','less1','never','4~8'],
    'less1': ['1~3','less1','never','4~8',np.nan],
    'never': ['never','1~3',np.nan]
}

def fill_restaurant20To50(row):
    if pd.isna(row['Restaurant20To50']) and row['RestaurantLessThan20'] in value_mapping:
        possible_values = [v for v in value_mapping[row['RestaurantLessThan20']] if pd.notna(v)]
        if possible_values:
            return np.random.choice(possible_values)
    return row['Restaurant20To50']
df_cl['Restaurant20To50']=df_cl.apply(fill_restaurant20To50, axis=1)
miss=df_cl.isnull().sum()
print(miss)
```

```python

df_cl['Restaurant20To50'] = df_cl['Restaurant20To50'].fillna(df_cl['Restaurant20To50'].mode()[0])
df_cl['RestaurantLessThan20'] = df_cl['RestaurantLessThan20'].fillna(df_cl['RestaurantLessThan20'].mode()[0])


miss=df_cl.isnull().sum()
miss
```
