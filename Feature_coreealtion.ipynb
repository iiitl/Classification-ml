```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv('/content/cleaned_dataset.csv')

# Display dataset information
df.info(), df.head()

```

```python
# Box plot for numerical features
numerical_features = ['temperature', 'toCoupon_GEQ5min', 'toCoupon_GEQ15min', 'toCoupon_GEQ25min']

plt.figure(figsize=(12, 8))
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='Y', y=feature, data=df)
    plt.title(f'Box plot of {feature} by Y')
plt.tight_layout()
plt.show()

```

```python
# Violin plot for categorical features
categorical_features = ['passanger', 'weather', 'coupon', 'gender', 'maritalStatus', 'education']

plt.figure(figsize=(12, 12))
for i, feature in enumerate(categorical_features):
    plt.subplot(3, 2, i+1)
    sns.violinplot(x='Y', y=df[feature].astype(str), data=df)
    plt.title(f'Violin plot of {feature} by Y')
plt.tight_layout()
plt.show()
```

```python

# Pairplot for numerical features
sns.pairplot(df, hue='Y', vars=numerical_features)
plt.show()
```

```python
# Heatmap of feature correlations (only numerical columns)
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()
```
