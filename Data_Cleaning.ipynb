# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Dataset

```python
data = pd.read_csv('in-vehicle-coupon-recommendation.csv')
data.head()
```

```python
data.info()
```

# Data Cleaning

## Removing Duplicates

```python
data.drop_duplicates(inplace=True)
```

```python
data.info()
```

### As the column car has 99.14% missing values its better to drop the column 

```python
data.drop('car', axis=1, inplace=True)
```

### Filling the Null values in object type columns Bar ,CoffeeHouse, CarryAway ,RestaurantLessThan20, Restaurant20To50 with their Mode.

```python
for col in ['Bar', 'CoffeeHouse', 'CarryAway', 'RestaurantLessThan20', 'Restaurant20To50']:
    data[col] = data[col].fillna(data[col].mode()[0])
```
