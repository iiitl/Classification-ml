```python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
for el in numerical_cols:
  Q1 = df[el].quantile(0.25)
  Q3 = df[el].quantile(0.75)
  IQR = Q3 - Q1

  df = df[~((df[el] < (Q1 - 1.5 * IQR)) | (df[el] > (Q3 + 1.5 * IQR)))]
```
