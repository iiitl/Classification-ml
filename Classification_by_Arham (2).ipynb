```python
import numpy as np
import pandas as pd
```

```python
recommendation=pd.read_csv("in-vehicle-coupon-recommendation.csv")
recommendation.head()
```

```python
#Checking all columns how many na values are there
recommendation.isna().sum()
```

```python
recommendation.nunique()
```

```python
recommendation.shape
```

```python
print(recommendation['car'].unique())
```

```python
#I am dropping the cars columns as it doesn't have a value most of the times
recommendation.drop(['car'],axis='columns',inplace=True)
```

```python
#Filling never for na values
recommendation['Bar']=recommendation['Bar'].fillna('never')
```

```python
# print(recommendation['CoffeeHouse'].unique())
# print(recommendation['CarryAway'].unique())
# print(recommendation['RestaurantLessThan20'].unique())
# print(recommendation['Restaurant20To50'].unique())
columns_to_fill = ['CoffeeHouse', 'CarryAway', 'RestaurantLessThan20', 'Restaurant20To50']
for column in columns_to_fill:
    recommendation[column] = recommendation[column].fillna('never')
```

```python
recommendation.head() #Reduced all na columns
#Data Cleaning is uptil here only
```

```python
from sklearn.preprocessing import LabelEncoder
lecol=['destination','passanger','weather','time','coupon','expiration','gender','age',
'maritalStatus','education','occupation','income','Bar','CoffeeHouse','CarryAway',
'RestaurantLessThan20','Restaurant20To50']
for col in recommendation[lecol]:
  le=LabelEncoder()
  recommendation[col]=le.fit_transform(recommendation[col])
# for col in lecol
#Using label encoding for objects
```

```python
recommendation.head()
```

```python
x=recommendation.drop('education',axis=1)
y=recommendation['education']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

Here we can see that using encoding we can improve the accuracy of the model

```python
model.score(x_train,y_train)
```
