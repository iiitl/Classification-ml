```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('/content/in-vehicle-coupon-recommendation.csv')
```

```python
df.head()
```

```python
df.drop(columns=['car','toCoupon_GEQ5min'])
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
df[['Bar','CoffeeHouse','CarryAway','RestaurantLessThan20','Restaurant20To50']] = imputer.fit_transform(df[['Bar','CoffeeHouse','CarryAway','RestaurantLessThan20','Restaurant20To50']])
```

```python
df.info()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title("Boxplot for Outliers in Multiple Columns")
plt.show()
```
